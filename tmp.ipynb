```python
import numpy as np
import pandas as pd

# 生成示例时间序列
g = 9.81
base_com = 1.0
crouch_depth = 0.25
crouch_duration = 0.35
push_duration = 0.12
v_takeoff = 3.0
flight_time = 2 * v_takeoff / g

t0 = 0
t_end = crouch_duration + push_duration + flight_time + 0.3
t = np.arange(t0, t_end, 0.005)

z = np.zeros_like(t)
phase = np.empty_like(t, dtype=object)

# crouch
idx_crouch = t < crouch_duration
z[idx_crouch] = base_com - (t[idx_crouch] / crouch_duration) * crouch_depth
phase[idx_crouch] = "crouch"

# push
idx_push = (t >= crouch_duration) & (t < crouch_duration + push_duration)
a_push = v_takeoff / push_duration
t_rel = t[idx_push] - crouch_duration
z[idx_push] = (base_com - crouch_depth) + 0.5 * a_push * t_rel**2
phase[idx_push] = "push"

# flight
idx_flight = (t >= crouch_duration + push_duration) & (t < crouch_duration + push_duration + flight_time)
t_rel = t[idx_flight] - (crouch_duration + push_duration)
z_takeoff = z[idx_push][-1]
z[idx_flight] = z_takeoff + v_takeoff * t_rel - 0.5 * g * t_rel**2
phase[idx_flight] = "flight"

# land
idx_land = t >= (crouch_duration + push_duration + flight_time)
t_rel = t[idx_land] - (crouch_duration + push_duration + flight_time)
z[idx_land] = base_com - 0.05 + 0.05 * np.exp(-6 * t_rel)
phase[idx_land] = "land"

df = pd.DataFrame({"time": t, "com_z": z, "phase": phase})
df.to_csv("jump_com_data.csv", index=False)

print("Saved:", "jump_com_data.csv")
```
